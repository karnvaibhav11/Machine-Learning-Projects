```python
from wordcloud import WordCloud, STOPWORDS
```

```python
import matplotlib.pyplot as plt
```

```python
def plotWC(wordC):
    plt.figure(figsize=(10,10))
    plt.imshow(wordC)
    plt.axis("off")
```

```python
processedText=WordCloud(width=500,height=500, background_color="white",random_state=7).generate(content)

plotWC(processedText)
```
